### 5.Создайте KDE (ядерная оценка плотности) для колонки 'defective'

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np


%matplotlib inline
sns.set(style="white")

# Plot the KDE for the "defective" column
plt.figure(figsize=(4,3), linewidth = 10000)
sns.kdeplot(df3['defective'], color='steelblue', linewidth=1.5)

plt.xlabel('', fontsize=0)
plt.ylabel('Density', fontsize=8)
plt.tick_params(axis='both', labelsize=8)
plt.xticks(fontsize=10)

min_value = df3['defective'].min()
max_value = df3['defective'].max()
custom_xticks = np.linspace(min_value, max_value, 6)
custom_xticks = np.round(custom_xticks).astype(int) 
plt.xticks(custom_xticks, fontsize=10)
plt.xlim(min_value - 2, max_value + 2)
plt.show()
```
